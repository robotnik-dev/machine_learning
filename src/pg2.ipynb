## `os.listdir`

```python
import os
```

```python
# list comprehension
l = []
for i in range(8):
    l.append(i**2)
    
l2 = [i**2 for i in range(8)]
print(l)
print(l2)
```

```python
p = "./data/MYON/train/"
labels = [f for f in os.listdir(p)]
paths = [p+f for f in os.listdir(p)]
print(labels)
print(paths)
```

```python
file_paths = [[(img_path, label) for img_path in os.listdir(p+label)] for label in os.listdir(p)]
print(file_paths)
```

## Magic und Generatoren

```python
def my_range(n):
    i = 0
    while i<n:
        yield i
        i += 1
```

```python
list(my_range(8))[4]
```

```python
l = list(range(9999))
s = set(range(9999))
```
